````
import numpy as np
import matplotlib.pyplot as plt
from sklearn.neighbors import KNeighborsClassifier as kNN
from sklearn.datasets import load_iris # Iris dataset
````

````
iris = load_iris() # Cargar la base de datos iris
iris_x = iris.data[:, :2]  # Seleccionar solo las dos primeras características
iris_y = iris.target # Etiquetas

# Imprimir formas
print (iris_x.shape, iris_y.shape)
````

````

from sklearn.datasets import load_iris
from sklearn.neighbors import KNeighborsClassifier  # Asegúrate de importar esta clase
# Instanciar el objeto kNN
knn = KNeighborsClassifier(n_neighbors=5, metric="euclidean")

# Entrenar el modelo kNN
knn.fit(iris_x, iris_y)

````

````
## Plot decision boundaries
# Valores mínimos y máximos para crear el grid de las fronteras de decisión
x_min, x_max = iris_x[:, 0].min() - 0.5, iris_x[:, 0].max() + 0.5
y_min, y_max = iris_x[:, 1].min() - 0.5, iris_x[:, 1].max() + 0.5

# Crear el grid de las fronteras de decisión
x_grid, y_grid = np.meshgrid(np.linspace(x_min, x_max, 100), np.linspace(y_min, y_max, 100))

# Predecir los valores del grid
grid_points = np.c_[x_grid.ravel(), y_grid.ravel()]
predictions = knn.predict(grid_points)
predictions = predictions.reshape(x_grid.shape)

# Visualizar la frontera de decisión
plt.figure(figsize=(8, 6))
plt.contourf(x_grid, y_grid, predictions, alpha=0.3, cmap=plt.cm.RdYlBu)
plt.scatter(iris_x[:, 0], iris_x[:, 1], c=iris_y, edgecolor='k', cmap=plt.cm.RdYlBu)
plt.xlabel('Largo del sépalo')
plt.ylabel('Ancho del sépalo')
plt.title(f'Frontera de decisión kNN (k={n_neighbors})')

# Agregar la barra de colores
cbar = plt.colorbar(label='Clase')
cbar.set_ticks([0, 1, 2])
cbar.set_ticklabels(['Setosa', 'Versicolor', 'Virginica'])

plt.show()
````

````

import itertools

# Cargar la base de datos Iris
iris = load_iris()
iris_x = iris.data  # Todas las características
iris_y = iris.target

# Configurar el número de vecinos
n_neighbors = 5

# Instanciar el objeto kNN
knn = KNeighborsClassifier(n_neighbors=n_neighbors, metric="euclidean")

# Nombres de las características
feature_names = [
    'Largo del sépalo',  # Feature 1
    'Ancho del sépalo',  # Feature 2
    'Largo del pétalo',  # Feature 3
    'Ancho del pétalo'   # Feature 4
]

# Crear gráficos de pares de características
for i, j in itertools.combinations(range(len(feature_names)), 2):
    # Seleccionar dos características para el gráfico
    iris_x_subset = iris_x[:, [i, j]]

    # Entrenar el modelo kNN con estas dos características
    knn.fit(iris_x_subset, iris_y)

    # Valores mínimos y máximos para crear el grid de las fronteras de decisión
    x_min, x_max = iris_x_subset[:, 0].min() - 0.5, iris_x_subset[:, 0].max() + 0.5
    y_min, y_max = iris_x_subset[:, 1].min() - 0.5, iris_x_subset[:, 1].max() + 0.5

    # Crear el grid de las fronteras de decisión
    x_grid, y_grid = np.meshgrid(np.linspace(x_min, x_max, 100), np.linspace(y_min, y_max, 100))

    # Predecir los valores del grid
    grid_points = np.c_[x_grid.ravel(), y_grid.ravel()]
    predictions = knn.predict(grid_points)
    predictions = predictions.reshape(x_grid.shape)

    # Visualizar la frontera de decisión
    plt.figure(figsize=(8, 6))
    plt.contourf(x_grid, y_grid, predictions, alpha=0.3, cmap=plt.cm.RdYlBu)
    plt.scatter(iris_x_subset[:, 0], iris_x_subset[:, 1], c=iris_y, edgecolor='k', cmap=plt.cm.RdYlBu)
    plt.xlabel(feature_names[i])
    plt.ylabel(feature_names[j])
    plt.title(f'Frontera de decisión kNN (k={n_neighbors}) - {feature_names[i]} vs {feature_names[j]}')

    # Agregar la barra de colores
    cbar = plt.colorbar(label='Clase')
    cbar.set_ticks([0, 1, 2])
    cbar.set_ticklabels(['Setosa', 'Versicolor', 'Virginica'])

    plt.show()
````

````

import itertools

features = ['mass', 'width', 'height', 'color_score']


# Nombres de las características y las clases
feature_names = features
class_names = ['Apple', 'Mandarin', 'Orange', 'Lemon']


# Crear gráficos de pares de características
for i, j in itertools.combinations(range(len(feature_names)), 2):
    # Seleccionar dos características para el gráfico
    X_subset = fruits_x[:, [i, j]]
    y_subset = fruits_y

    # Entrenar el modelo kNN con estas dos características
    knn.fit(X_subset, y_subset)

    # Valores mínimos y máximos para crear el grid de las fronteras de decisión
    x_min, x_max = X_subset[:, 0].min() - 0.5, X_subset[:, 0].max() + 0.5
    y_min, y_max = X_subset[:, 1].min() - 0.5, X_subset[:, 1].max() + 0.5

    # Crear el grid de las fronteras de decisión
    x_grid, y_grid = np.meshgrid(np.linspace(x_min, x_max, 100), np.linspace(y_min, y_max, 100))

    # Predecir los valores del grid
    grid_points = np.c_[x_grid.ravel(), y_grid.ravel()]
    predictions = knn.predict(grid_points)
    predictions = predictions.reshape(x_grid.shape)

    # Visualizar la frontera de decisión
    plt.figure(figsize=(8, 6))
    plt.contourf(x_grid, y_grid, predictions, alpha=0.3, cmap=plt.cm.RdYlBu)
    plt.scatter(X_subset[:, 0], X_subset[:, 1], c=y_subset, edgecolor='k', cmap=plt.cm.RdYlBu)
    plt.xlabel(feature_names[i])
    plt.ylabel(feature_names[j])
    plt.title(f'Frontera de decisión todas caracteristicas kNN (k={n_neighbors}) - {feature_names[i]} vs {feature_names[j]}')

    # Agregar la barra de colores
    cbar = plt.colorbar(label='Clase')
    cbar.set_ticks([0, 1, 2, 3])
    cbar.set_ticklabels(class_names)

    plt.show()
````
